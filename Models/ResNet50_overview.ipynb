```python
# ResNet50

import torch
import torch.nn as nn
from torchvision import models
```

```python
from torchvision.models.resnet import model_urls
model_urls['resnet50'] = model_urls['resnet50'].replace('https://', 'http://')
resnet50model = models.resnet50(pretrained = True)
```

```python
resnet50model
```

```python
resnet50model.fc
```

```python
#Linear Layer

resetlayer = resnet50model.fc

# extract parameters
in_channels = resetlayer.in_features
out_channels = resetlayer.out_features
bias = resetlayer.bias


num_channels = 27
```

```python

# create new layer
resnet50model.fc = nn.Linear(in_channels, num_channels)
```

```python
resnet50model
```

```python
nn.Sequential(*list(resnet50model.children())[:-1]) # remove dense_block 4 and following: [:-2]
```

```python
features = nn.Sequential(*list(resnet50model.children())[:-1]) # remove dense_block 4 and following: [:-2]
classifier = nn.Linear(in_channels, num_channels)
```

```python
nn.Sequential(features, classifier)
```
